**1. Summarize the assumptions of a multiple regression model.**

The assumptions of the multiple regression model are the following:

* **Linear Relationship**: There is a linear relationship between the dependent and independent variables
* **No Multicollinearity**: The independent variables are not highly correlated with one another
* **Independence**: The response variable (Yi) observations are selected randomly from the population
* **Homoscedasticity**:The residuals have constant variance at every point in the linear model.
* **Multivariate Normality**: The residuals are normally distributed with a mean of zero.

(Bhandari, 2020)

```python
# We see in the heat map that DummyStroke, DummyEmergency_admin and DummyHighComplication_risk have 
# a significantly lower correlation compared to Age and DummyHighBlood
# Let's run an OLS on just Age and DummyHighBlood

df['intercept'] = 1

lm_bandwidth_reduced = sm.OLS(df['Additional_charges'], df[['Age', 'DummyHighBlood', 'intercept']]).fit()

print(lm_bandwidth_reduced.summary())
```

We have reduced 33 variables to 2, with a r-squared value of 0.935 indicating that these two variables explain 93.5% of the variance. This is in comparison to the 33 variables that explained 93.9% of the variance. The p-values of 0 indicates that the variables are significant (Bhandari, 2020).

We will see the reduced equation below.

Our reduced model is:

**y = 225.7781 * Age + 8639.1132 * DummyHighBlood**

With simply the age of the individual and whether they have high blood pressure or not, we can predict how much their additional charges will be.
